```python
a = df[['lat','lng']].reset_index()
del a['index']
a.head()
```

```python
plotArr = pd.DataFrame(plotArr)
plotArr.head()
```

```python
b = a.join(plotArr)
b.head()
```

```python
def createColors(reds, greens, blue):
    clr_tups = zip(mapRange(reds,1), mapRange(greens, 1), blue*np.ones(len(reds)))
    return [mclr.rgb2hex(rgb) for rgb in clr_tups]
```

```python
def mapRange(arr,rg):
    return (arr - np.min(arr)) / (np.max(arr) - np.min(arr)) * rg
#mapRange(b['lat'],1)
```

```python
coorColors = createColors(b['lat'], b['lng'], 0.5)
#area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radiuses
plt.figure(figsize=(16,10))
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.xlim([-0.7,0.5])
plt.ylim([-0.6,0.7])
plt.scatter(b.loc[:,1], b.loc[:,0], c=coorColors, alpha=0.8, s=15, edgecolors='none')
plt.show()
```
