```python
import requests
import re
import time
```

```python
urls = ['http://www.yelp.com/biz/salt-and-straw-los-angeles',
        'http://www.yelp.com/biz/smittys-famous-fish-and-chicken-culver-city',
        'http://www.yelp.com/biz/zankou-chicken-los-angeles-7?osq=Zankou+Chicken',
        'http://www.yelp.com/biz/chego-los-angeles-5',
        'http://www.yelp.com/biz/ambala-dhaba-homestyle-indian-los-angeles',
        'http://www.yelp.com/biz/the-colony-cafe-los-angeles']

req = requests.get(urls[1])
rtext = req.text
```

```python
for url in urls:            # '(?<=<span class="category-str-list">)[^>]*(?<=>)([^<>]*)(?=<)[^<]*(?=<\span>)'
    req = requests.get(url) #'<span\s[^>]*class="category-str-list"[^>]*>[\s]*<a[^<>]*>([^<>]*)<\/a>'
    rtext = req.text        #'<span class="category-str-list">[\s]*<a[^<>]*>([^<>]*)<\/a>'
    result = re.search(r'<span\s[^>]*itemprop="addressLocality"[^>]*>([^<>]*)[\s]*<\/span>', rtext)
    #time.sleep(500)
    print result.group(1)
                    
```
